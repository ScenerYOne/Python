1.   run code 

```python
import numpy as np
from tensorflow.keras.datasets.mnist import load_data

(X_train, y_train), (X_test, y_test) = load_data()
```

```python
num_classes = 10
```

```python
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], X_train.shape[2], 1))
```

```python
X_test = X_test.reshape((X_test.shape[0], X_test.shape[1], X_test.shape[2], 1))
```

```python
in_shape = X_train.shape[1:]
in_shape
```

2.   run code Reshape?


ans

3.  model.summary

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Conv2D, MaxPool2D, Flatten, Dropout
```

```python
model = Sequential()
model.add(Conv2D(32, (3,3), activation='relu', input_shape=in_shape))
model.add(MaxPool2D((2, 2)))
```

```python
model.add(Conv2D(64, (3,3), activation='relu'))
model.add(MaxPool2D((2,2)))
```

```python
model.add(Flatten())
model.add(Dropout(0.5))
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
```

```python
model.add(Dense(num_classes, activation='softmax'))
```

```python
model.summary()
```

4.   plot model

```python
from tensorflow.keras.utils import plot_model
plot_model(model, 'digital.png', show_shapes = True, show_layer_names=True)
```

5.   Neural Network

```python
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
import time
start = time.time()

history = model.fit(X_train, y_train, epochs=15, batch_size=128, verbose=1, validation_split=0.1)
end = time.time()
print("Time taken: {:.2f} minutes".format((end-start)/60))
```

6.  run code 

```python
score = model.evaluate(X_test, y_test, verbose=1)
print('Accuracy: {:3f}', format(score[1]))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(history.history['loss'], color='b', label='Traning')
plt.plot(history.history['val_loss'], 'r--', label='validation')
plt.title('Loss')
```

7.  run code 

```python
y_pred = model.predict(X_test)
y_pred_cls = np.argmax(y_pred, axis=1)
idx_miss = np.where(y_pred_cls != y_test)
print(len(idx_miss[0]))
print(idx_miss[0][:15])
```
