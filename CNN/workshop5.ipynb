```python
def visual_multi(i):
  nplots = 40
  fig = plt.figure(figsize=(8, 4))
  for j in range(nplots):
    plt.subplot(4, 10, j+1)
    plt.imshow(x_train[i+j], cmap=plt.cm.gray_r)
    plt.title(y_train[i+j])
    plt.xticks([]); plt.yticks([])
  plt.show
```

```python
visual_multi(0)
```
